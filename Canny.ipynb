```python
import cv2
from util import imread, Canny, show_im
```

```python
path = ""
i_name = "book"
ext = "jpg"
    
im = imread(path, i_name, ext)

intermediate_images = ["X", "Y", "NMS", "Threshold"]

edges, imgs, titles = Canny(im, intermediate_images)

#show intermediate images
show_im(imgs, titles)
```

```python
cv2_edges = cv2.Canny(image = im, threshold1=100, threshold2=200, apertureSize=3, L2gradient=False)

show_im([edges, cv2_edges], ["My Canny Impl: ", "CV2 Canny"])
```

```python
"""Optional: Files can be saved here"""
cv2.imwrite(path + i_name + "_result.jpg", edges) 
for i in range(len(imgs)):
    img = imgs[i]
    title = titles[i]
    cv2.imwrite(path + i_name + "_" + title + ".jpg", img) 
```
